```python
# file_path = Path(__file__).parent.joinpath('Files', 'TSLA_quote.csv')
file_path = '/content/gdrive/MyDrive/Colab Notebooks/Files/TSLA_quote.csv'
df_tesla = pd.read_csv(file_path)
df_tesla['Date'] = df_tesla['Date'].apply(pd.to_datetime)
df_tesla.head()
```

```python
start_date = dt.datetime(2017,1,20)
end_date = dt.datetime(2023,11,30)
```

```python
df_tesla.loc[df_tesla.Date==start_date, 'Close/Last']
```

```python
start_price = df_tesla.loc[df_tesla.Date==start_date, 'Close/Last'].item()
start_price = float(start_price.replace('$',''))
print(start_price)
```

```python
end_price = df_tesla.loc[df_tesla.Date==end_date, 'Close/Last'].item()
end_price = float(end_price.replace('$',''))
print(end_price)
```
