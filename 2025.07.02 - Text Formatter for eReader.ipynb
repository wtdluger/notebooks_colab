This notebook takes an input text file with various line breaks, and creates a new text file with each paragraph on a single line of text.

```python
from pathlib import Path
import datetime as dt
```

```python
# If remounting (i.e. to reload updated data), run cell twice
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
```

```python
directory = Path('/content/gdrive/MyDrive/Colab Notebooks/Files/')
# file_name = "Rusty Schweickart - No Frames No Boundries.txt"
file_name = 'JFK - We Choose to go to the Moon.txt'

file_path = directory.joinpath(file_name)
```

```python
output_name = (file_name.split('.txt')[0] + ' '
               + dt.datetime.now().strftime("%Y-%m-%d_%H-%M") + '.txt'
)
output_path = directory.joinpath(output_name)
print(output_path)
```

```python
with open(file_path, 'r') as file_handler:
    file_string = file_handler.read()
```

```python
# Convert sentance spacing from old style double space to modern single space.
# For modernity and to aid with formatting code below.
new_string = file_string.replace('.  ', '. ')

# make an entire paragraph a single line of text
# convert end of line ' \n' to a single space.
new_string = new_string.replace(' \n', ' ')

# separate paragraphs with '\n \n'
new_string = new_string.replace('\n', ' ')
new_string = new_string.replace('  ', '\n \n')
```

```python
new_string
```

```python
# save modified string as a text file with original name + time stamp appended
# to the same directory as original file
with open(output_path, 'w', encoding='utf-8') as file_handler:
    file_handler.write(new_string)
```
