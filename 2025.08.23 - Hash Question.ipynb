```python
import json
import hashlib
import requests
```

```python
url = 'https://api.close.com/buildwithus/'
```

GET Request

```python
resp1 = requests.get(url)
```

```python
resp1.text
```

```python
traits = ["Craftsman", "Pragmatic", "Curious", "Methodical", "Driven", "Collaborator"]
key = "Close-2bf739ef"
```

```python
hashes = []
for trait in traits:
    hash = hashlib.blake2b(trait.encode("utf-8"), digest_size=64, key=key.encode("utf-8")).hexdigest()
    hashes.append(hash)
```

```python
hashes
```

```python
json.dumps(hashes)
```

POST Request for Verification Code

```python
resp2 = requests.post(url, json=hashes)
```
