```python
import numpy as np
import time
from IPython.display import Audio, display
```

```python
def timer(t, message="Time's up!"):
    """
    Function Takes second input and pauses program
    for that amount of time

    Optional parameter:
        message: default "Time's up!"
        string to print upon conclusion of the function
        if message == None, nothing will be printed
    """
    while t:
        mins, secs = divmod(t, 60)
        timer = '{:02d}:{:02d}'.format(mins, secs)
        print(timer, end="\r")  # Overwrite previous output
        time.sleep(1)
        t -= 1
    if (message != None):
        print(message)
```

```python
# Generate a simple sine wave
sample_rate = 44100
duration = 1  # seconds
frequency = 196  # 440 Hz (A4 note)
t = np.linspace(0., duration, int(sample_rate * duration))
amplitude = np.iinfo(np.int16).max * 0.5
data = amplitude * np.sin(2. * np.pi * frequency * t)
```

```python
# pomodoro timing inputs

# pomodoro intervals
pomodoros = 5

# work duration, minutes component
work_min = 0
# work duration, seconds component
work_s = 3

# break duration, minutes component
break_min = 0
# break duration, seconds component
break_s = 1

# convert work and break durations into seconds
work_s += 60*work_min
break_s += 60*break_min
```

```python
# Pomodoro Timer Proper
print("Begin Work!")
display(Audio(data, rate=sample_rate, autoplay=True))
for p in range(2, pomodoros+1):
    timer(work_s, "Take a Break!")
    display(Audio(data, rate=sample_rate, autoplay=True))
    timer(break_s, "Begin Work!")
    display(Audio(data, rate=sample_rate, autoplay=True))
timer(work_s, "You're Finished! Congratulations!")
display(Audio(data, rate=sample_rate, autoplay=True))
```
