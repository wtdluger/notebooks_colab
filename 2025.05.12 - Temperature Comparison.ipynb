```python
import numpy as np
from scipy import special
import sympy as sp
from matplotlib import pyplot as plt
```

# Conversion Functions

```python
# nPR: 3P2
# 3 Common Temperature Units to convert from/to
special.perm(3,2)
# there should be 6 conversion functions
```

```python
# Temperature Conversions
def f2c(f):
    c = (f-32)*5/9
    return c
def c2k(c):
    k = c+273.15
    return k
def c2f(c):
    f = c*9/5+32
    return f
def k2c(k):
    c = k-273.15
    return c
def k2f(k):
    f = c2f(k2c(k))
    return f
def f2k(f):
    k = c2k(f2c(f))
    return k
```

# When does the Celsius Temperature Equal Fahrenheit Temperature?

```python
# Range of Temperatures to Compare
t = np.arange(-100,100)

# Celsius values to Kelvin
c = c2k(t)

# Fahrenheit values to Kelvin
f = f2k(t)
```

```python
t[f==c]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, dpi=100)
axes.plot(t, c, label='Celsius')
axes.plot(t, f, label='Fahrenheit')
axes.axvline(x=-40, color='r', ls='--', label='-40Â°')
axes.legend(loc='best')
axes.grid(True)
```

# When does the Kelvin Temperature Equal Fahrenheit Temperature?

```python
# Range of Temperatures to Compare
t = np.arange(500,700)

# Fahrenheit values to Kelvin
f = f2k(t)
```

```python
kelvin = sp.symbols('T_{k}')
fahrenheit = sp.symbols('T_f')

fahrenheit_eq = sp.Eq(kelvin, f2k(kelvin))
fahrenheit_eq
```

```python
sp.solve(fahrenheit_eq, kelvin)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, dpi=100)
axes.plot(t, t, label='Kelvin')
axes.plot(t, f, label='Fahrenheit')
axes.axvline(x=574.5875, color='r', ls='--', label='574.5875')
axes.legend(loc='best')
axes.grid(True)
```
